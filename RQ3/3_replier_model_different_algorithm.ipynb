#### **Different Algorithms**

```python
import helper.stat_helper as stat_hp
importlib.reload(stat_hp)

def test_balanced(df_stat, algo_list, filename=None):
    '''
    Runs 10 fold cross validation in 10 different dataset with
    different algorithm
    :param df_stat: Dataframe with features
    :param alo_list: List of algorithms to try
    :param filename: Filename to save the result

    '''
    all_cv_scores = []
    for algo in algo_list:
        df_0 = df_stat.loc[df_stat['replier_label'] == 0]
        df_1 = df_stat.loc[df_stat['replier_label'] == 1]

        #### Run test on 10 different samples
        #### sample without replacement
        for i in range(0,10):
            df_sample = df_0.sample(len(df_1), random_state=i)

            df_0 = df_0.loc[~df_0['replier_userid'].isin(
                df_sample['replier_userid']
            )]

            df_all = pd.concat([df_sample,
                                df_1
                               ])

            cv_results_tuned_model = \
            stat_hp.run_model_with_best_threshold(
                df_all,
                columns_not_include=['list_age'],
                model_type=algo, 
                y_column = 'replier_label',
                filename=None,
            )

            all_cv_scores.append(cv_results_tuned_model)

    (pd.concat(all_cv_scores)
    ).to_pickle(filename)
```

```python
filename ='./../results/replier_classifier_different_algorithm.pkl.gz'

algo_list = ['logistic', 'random', 'ada', 'tree', 'naive']

importlib.reload(stat_hp)

test_balanced(df_stat, algo_list, filename=filename)
```
