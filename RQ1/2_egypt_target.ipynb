#### **This notsbook prepares the exploratory plots of manually annotated data**

```python
import pandas as pd

import helper.strategy_helper as st_hp
import helper.visualization as vz_hp
import config.config as config_hp
```

```python
df = pd.read_csv('./../data/RQ1_target_annotation.csv')
```

```python
df_egypt = df.loc[df['campaign'] == 'Egypt']
```

#### **Egypt target country visualization**

```python
import importlib

importlib.reload(config_hp)

df_egypt.loc[df_egypt['country_label'] == 'Saudi Arabia', 
                   'country_label'] = 'Saudi\n Arabia'

df_country_grp = (df_egypt.groupby(['country_label'])
                  .size()
                  .to_frame('count')
                  .reset_index()
                 )

df_country_grp = df_country_grp.sort_values(by=['count'],
                                            ascending=False
                                           ).head(10)

x = df_country_grp['country_label'].tolist()
y = df_country_grp['count'].tolist()

parameters =  {
        'size': (10, 10),
        'fontsize': 36,
        'tick_size': 28,
        'complementary': True,
        'bar': 'h',
        'x': x,
        'y': y,
        'xlabel': 'Number of targets', 
        'ylabel': 'Country',
        'save': {
            'path': './../plots',
            'filename': 'egypt_022020_target_country_top_10.png'
        },
        'random_color': False
    }

vz_hp.bar_graph(parameters)
```

#### **Egypt target profession visualization**

```python
df_person_grp = (df_egypt.groupby(['profession_label'])
                  .size()
                  .to_frame('count')
                  .reset_index()
                 )


df_person_grp = df_person_grp.sort_values(by=['count'],
                                          ascending=False
                                         ).head(10)

x = df_person_grp['profession_label'].tolist()
y = df_person_grp['count'].tolist()

parameters =  {
        'size': (10, 10),
        'fontsize': 36,
        'tick_size': 28,
        'complementary': True,
        'bar': 'h',
        'x': x,
        'y': y,
        'xlabel': 'Number of targets', 
        'ylabel': 'Profession',
        'save': {
            'path': './../plots',
            'filename': 'egypt_022020_target_profession_top_10.png'
        },
        'random_color': False
    }



vz_hp.bar_graph(parameters)
```
