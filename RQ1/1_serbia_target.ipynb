#### **This notebook preprocess, visualize the targets and their profession for serbia campaigns**

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import matplotlib.colors as pltc

import importlib

#### packages
import helper.visualization as vz_hp
import config.config as config_hp
```

#### **Load file**

```python
df = pd.read_csv('./../data/RQ1_target_annotation.csv')
```

```python
df_serbia = df.loc[df['campaign'] == 'Serbia']
```

#### **Serbia target country visualization**

```python
importlib.reload(config_hp)
importlib.reload(vz_hp)

df_country_grp = (df_serbia.groupby(['country_label'])
                  .size()
                  .to_frame('count')
                  .reset_index()
                 )

df_country_grp = df_country_grp.sort_values(by=['count'],
                                          ascending=False
                                           ).head(10)

df_country_grp['str_length'] = df_country_grp['country_label'].apply(
    lambda x: len(x)
)

x = df_country_grp['country_label'].tolist()
y = df_country_grp['count'].tolist()

parameters =  {
        'size': (10, 10),
        'fontsize': 36,
        'tick_size': 28,
        'complementary': True,
        'bar': 'h',
        'x': x,
        'y': y,
        'xlabel': 'Number of targets', 
        'ylabel': 'Country',
        'save': {
            'path': './../plots',
            'filename': 'serbia_022020_target_country_top_10.png'
        },
        'random_color': False
    }

vz_hp.bar_graph(parameters)
```
